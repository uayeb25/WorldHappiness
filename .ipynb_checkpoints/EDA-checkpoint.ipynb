## Ejercicios

1. Partiendo de Wh, realice un analisis de valores nulos. que porcentaje de valores nulos se encuentran en cada columna por anio. presente el analisis en un dataframe

Ejemplo:

```python
pd.DataFrame({
    "year": [2017,2017]
    , "column" : ["Happiness.Rank","Whisker.low"]
    , "percentage.of.nans" : [0.2,0.2]
})
```

### resolucion ejercicio uno

```python
result = pd.DataFrame()
```

```python
for year_ in wh.loc[ : , "year" ].drop_duplicates():
    wh_year = wh.loc[ wh['year'] == year_ , : ]    
    for column_ in wh_year.columns:        
        percentage =  np.sum(pd.isna( wh_year.loc[ : , column_ ]  )) / wh_year.shape[0]
        dfrow = pd.DataFrame({
            'year': [ year_ ]
            , 'column' : [ column_ ]
            , 'percentage.of.nans' : [ percentage ]
        })
        result = pd.concat([result, dfrow])
```

```python
result
```

3. partiendo de Wh, remplace los valores NaN de las columnas por su mean, el mean debe de ser calculado por año para ser mas exacto

### resolucion ejercicio tres

```python
wh.columns
```

```python
columns_to_use = ['Score', 'GDP per capita',
       'Social support', 'Healthy life expectancy',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']
```

```python
wh_limpio  = pd.DataFrame()
for year_ in wh.loc[ : , "year" ].drop_duplicates():
    wh_year = wh.loc[ wh['year'] == year_ , : ].copy()
    print( "Working with: {0}".format(year_) )
    for c_ in columns_to_use:        
        mean_column = np.mean( wh_year.loc[ : , c_ ] )
        print( "Column: {0} has an mean of {1}".format(c_,mean_column) )        
        wh_year.loc[ np.isnan( wh_year[c_] ) , c_ ] = mean_column
    
    wh_limpio = pd.concat([ wh_limpio.copy() , wh_year.copy() ])
```

```python
wh_limpio.loc[ np.isnan( wh_limpio['Score'] ) , : ]
```

```python
wh.loc[ np.isnan( wh['Score'] ) , : ]
```

```python
x = wh_limpio.sort_values(["Country or region","year"] , ascending = False )
x["rank"] = x.groupby(["Country or region"])["year"].rank( method = 'first' , ascending = False )
x = x.loc[ x["rank"] == 1.0 , : ].reset_index(drop=True)
```

```python
y = wh_limpio.sort_values(["Country or region","year"] , ascending = True )
y["rank"] = y.groupby(["Country or region"])["year"].rank( method = 'first' , ascending = True )
y = y.loc[ y["rank"] == 1.0 , : ].sort_values(["Country or region"] , ascending = False ).reset_index(drop=True)
```
